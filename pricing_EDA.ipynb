Se analisarmos a tipo da coluna last_review com cuidado, vemos que ela é do tipo object, porém ela guarda a data da última review, logo vamos transformá-la em uma datetime.

```python
df['last_review'] = pd.to_datetime(df['last_review'])
df.last_review
```

Pronto! Agora se notarmos, antes, para valores nulos, ela retornava um NaN e agora está retornando um NaT (Not a Time).

Após verificarmos o tipo das features e transformar a coluna anterior. Vamos setar uma variável com as features numéricas e a feature temporal (last_review) que será bastante usada posteriormente.

```python
numeric_features = ['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
                    'host_listing_count_calculated', '365_days_availability', 'last_review']
```

Para tomar consciência da quantidade de campos nulos e da porcentagem, resolvi criar a tabela a seguir

```python
sum_null_data = df.isnull().sum()
missing_data = sum_null_data.to_frame('sum').join((sum_null_data / df.shape[0] * 100).to_frame('percentage'))
missing_data
```

Ao analisar a tabela anterior, vemos que as colunas last_reviews e reviews_per_month têm uma porcentagem de 20.6% de nulos. Para saber o impacto das colunas last_review e reviews_per_month, vamos plotar uma matriz de correlação e analisar o grau de correlação dessas colunas com a coluna price (nossa coluna target).

```python
correlation_matrix = df[numeric_features].corr()
plt.figure(figsize=(10,10))
plt.title('Matriz de Correlação')

sns.heatmap(correlation_matrix, annot=True, cmap="copper")
```

Ao analisar o gráfico anterior, é possível observar que as colunas last_review e review_per_month têm uma correlação negativa com a coluna price, somando isso com a taxa de 20% de dados faltantes e também para não ter um impacto de dados faltantes na nossa análise, vamos remover essas colunas do dataset já que elas não interferem no preço do aluguel do imóvel.

```python
df = df.drop(columns=['last_review', 'reviews_per_month'])
numeric_features.remove('last_review')
numeric_features.remove('reviews_per_month')
```

Na célula a seguir, um gráfico KDE é plotado com o objetivo de analisar a distribuição contínua das features numéricas.

```python
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=((12, 8)))
axes=axes.flatten()
plt.suptitle('Distribuição da Densidade dos Dados Numéricos')

for i, feature in enumerate(numeric_features):
    sns.kdeplot(x=df[feature], fill=True, ax=axes[i])
    axes[i].set_xlabel(feature)

plt.tight_layout()
```

Com a plotagem do gráfico, ficou possível visualizar onde a moda de cada feature se concentra (a maior altura nas curvas), como os dados estão dispersos (se a curva for mais estreita, os dados estão mais concentrados próximos uns dos outros) e a assimetria. A assimetria da maioria dos gráficos ficou sendo à direita, o que significa que os dados se estendem para valores mais altos, podendo até ser possíveis outliers. Vamos usar o .describe() para obtermos mais informações:

```python
df.describe()
```

Para compreender rapidamente a distribuição, usamos o .describe() para ver medidas importantes como a média (mean), desvio padrão (std) quartis (25%, 50%, 75%) e os valores máximos e mínimos. Vamos analisar a coluna price, o nosso target, vemos que a média fica em torno dos 152 dólares/noite, o desvio padrão está bem alto, o que significa que com um desvio de 240 dólares, os preços variam bastante em relação à média. Podemos ver que 75% dos valores estão a abaixo de 175 dólares, mas vemos também que o valor máximo é 10.000 dólares, o que significa que temos outliers. Vamos plotar um gráfico de caixa para termos a certeza dos outliers dessa coluna:

```python
fig, axes = plt.subplots(1, figsize=((12, 8)))
plt.title('Distribuição dos Preços')

for i, feature in enumerate(['price']):
    sns.boxplot(y=df[df[feature]<1000][feature])
    plt.xlabel('price')

plt.tight_layout()
```

Pode-se ver que a coluna price tem muitos outliers e o valor típico máximo não ultrapassa 400 dólares/noite (chamado de bigode). Vamos remover os outliers da coluna preço e ver se influencia em algum resultado:

```python
df_cleaned = df[df['price'].between(0, 280)]
df_cleaned.describe()
```

```python
fig, axes = plt.subplots(1, figsize=((12, 8)))
plt.title('Distribuição dos Preços')

for i, feature in enumerate(['price']):
    sns.boxplot(y=df_cleaned[df_cleaned[feature]<1000][feature])
    plt.xlabel('price')

plt.tight_layout()
```

Removendo os outliers da coluna price, vimos que os resultados mudaram consideravelmente, assim a média diminuindo de 152 para 113 e o terceiro quartil diminuindo de 175 para 150. Mas como esse outliers não são dados errados, porque claramente uma casa pode valer 10000 doláres/noite, iremos continuar manipuladando os dados com outliers.

Para uma análise univariada, vamos ver a popularidade de cada grupo de bairros e cada tipo de acomodação.

### 1. Supondo que uma pessoa esteja pensando em investir em um apartamento para alugar na plataforma, onde seria mais indicada a compra?

Ao alugar um imóvel, é importante ter conhecimento das seguintes coisas: localização, preço e disponibilidade no ano. Para atender os seguintes tópicos, iremos supor que uma boa localização é onde mais tem imóveis para alugar, com o preço menor que 175 dólares/noite (75% do valor dos imóveis estão abaixo desse preço) e disponibilidade de 365 dias por ano.

```python
filtered_df = df[(df['price']<=175)&(df['365_days_availability']==365)]
```

Para saber quais bairros mais tem imóveis para comprar, vamos plotar uma tabela e vamos pegar os 5 mais populares bairros:

```python
popular_neighborhood = df.neighborhood.value_counts().head(5)
popular_neighborhood = popular_neighborhood.index.tolist()
```

```python
filtered_df = df[df['neighborhood'].isin(popular_neighborhood)]
```

Agora vamos agrupar as linhas por bairro e depois selecionar a coluna de preço para calcularmos a média de preços em cada bairro. Depois botamos em ordem crescente.

```python
result=filtered_df.groupby('neighborhood')['price'].agg('mean').sort_values(ascending=True)
result
```

Para melhor visualização, vamos plotar um gráfico para ver o preço médio para cada bairro.

```python
plt.figure(figsize=(10,10))
plt.title('Média de Preço por Bairro')
plt.xlabel('Bairro')
plt.ylabel('Preço')

sns.barplot(data=result.reset_index(), x='neighborhood', y='price')
```

Logo, a indicação seria comprar uma casa em Bushwick, com a média do aluguel sendo de 84,80 doláres/noite.

### 2. O número mínimo de noites e a disponibilidade ao longo do ano interferem no preço?

Para resolver esse mistério, vamos tentar correlacionar as três variáveis numéricas, plotando uma matriz de correlação.

```python
correlation_matrix = df[numeric_features].corr()
plt.figure(figsize=(10,10))
plt.title('Matriz de Correlação')

sns.heatmap(correlation_matrix, annot=True, cmap="copper")
```

Ao analisar a matriz de correlação, vemos que a correlação entre a coluna price e a coluna minimum_nights é de 0.043 e entre price e 365_days_availability é de 0.082. Isso demonstra que nenhuma coluna tem uma interferência tão grande na questão do preço, já que elas não estão tão correlacionadas. Para melhor visualização do impacto delas, vamos plotar dois gráficos para analisar:

```python
features = ['minimum_nights', '365_days_availability']
fig, axes = plt.subplots(1, 2, figsize=((12, 8)))
axes=axes.flatten()

for i, feature in enumerate(features):
    sns.scatterplot(data=df, x=features[i], y='price', ax=axes[i])
    plt.xlabel(categorical_features[i])
    axes[i].set_xlabel(feature)

plt.tight_layout()
```

No gráfico de mínimo de noites, conseguimos ver uma relação entre a quantidade de noites e o preço, no caso, quanto maior a quantidade mínima de noites, menor o preço, mas isso não é forte o suficiente para interferir no preço, pois quando os valores de mínimo de noites são pequenos, os preços variam bastante. Já no gráfico ao lado, a relação entre eles ficam bem menos clara, pois os preços estão bastante distribuídos na quantidade de dias disponível. Com isso, concluímos que não interfere.

### 3. Existe algum padrão no texto do nome do local para lugares de mais alto valor?

Para vermos se tem padrão, vamos plotar uma nuvem de palavras para imóveis acima de 280 e abaixo de 280. Escolhi 280 porque nas análises anteriores vimos que os preços se concentravam mais até 280, sendo esse valor o 'bigode' do gráfico boxplot.

```python
m = df[df['price']>=280]
wc = WordCloud().generate(str(m['name'].values))
plt.imshow(wc, interpolation='bilinear')
```

Podemos ver que as palavras que se destacam mais são 'Spacious', 'Large', 'Sweet', 'Luxury', os quais são adjetivos que tem uma entonação de ostentação e 'Brooklyn', o qual está no segundo lugar de bairros mais caros. Vemos também as palavras 'spa', 'massage', 'penthouse', 'garden', 'bbq', as quais são ambientes que encarecem os imóveis.

```python
m = df[df['price']<280]
wc = WordCloud().generate(str(m['name'].values))
plt.imshow(wc, interpolation='bilinear')
```

E nesse outro gráfico se destacam palavras como: 'cozy', 'kitchen', 'skylit', 'village', que são comuns em todas as casas e não vemos adjetivos tão ostensivos, apesar da palavras 'Castle', 'Harlem' e 'New York', não sendo bairros tão baratos.

Assim, podemos concluir que existe sim um padrão. O padrão é a forma como o nome do imóvel é anunciado, sendo os mais caros com mais adjetivos qualitativos e os mais baratos com palavras mais descritivas quanto ao local.
